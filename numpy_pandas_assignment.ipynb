```python
# c. creating 1-D array called arry1 which contains elements ranging from 0 to 27 

arry1 = np.arange(28)
print(arry1)
```

```python
# ci. converting arry1 to 3-D array called arry2

arry2 = arry1.reshape(2,2,7)
print(arry2)
```

```python
# di. creating two 2-D arrays with one ranging from 1 to 6 and the other 7 to 12

array_1 = np.arange(1,7).reshape(2,3)
array_2 = np.arange(7,13).reshape(2,3)
print(array_1)
print(array_2)
```

```python
#dii. horizontally and vertically stacking the two arrays together

horizontal_stack = np.hstack((array_1,array_2))
vertical_stack = np.vstack((array_1,array_2))
print("Horizontal Stack")
print(horizontal_stack)
print("Vertical Stack")
print(vertical_stack)
```
