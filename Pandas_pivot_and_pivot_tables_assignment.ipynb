```python
import pandas as pd
```

```python
# importing the autos.csv dataset
df = pd.read_csv("autos.csv")
```

```python
df
```

```python
# printing out the first 8 entries of the dataset

df.head(8)
```

```python
# 3. A DataFrame containing the minimum price of each car 
    
df.groupby('make')['price'].min().reset_index()
```

```python
# 4. Returning the smallest standard deviation of the prices

df.groupby('make')['price'].std().min()
```

```python
# 5. Returning the last two rows of each group of car make and fuel_type

df.groupby(['make','fuel_type']).mean().tail(2)
```

```python
# 6. Replicating the split-apply-combine without groupby
# split-apply-combine involves splitting the data into groupes,applying a particular function and later combine.
# will the dataset, I will split into make and fuel_type and apply the aggregate function to find;
# maximum, minimum and mean price.

df_group = df.groupby(['make','fuel_type'])
df_agg_price = df_group['price'].agg(['max','min','mean'])
df_agg_price
```

```python
# 7. Creating a separate dictionaries and merging them as a new dataset
# First dictionary
dict_1 = {"make":df.make, "price":df.price, "fuel_type":df.fuel_type}
df1 = pd.DataFrame(dict_1)
df1
```

```python
# Second dictionary
dict_2 = {"num_of_doors":df.num_of_doors, "engine_size":df.engine_size, "wheel_base":df.wheel_base}
df2 = pd.DataFrame(dict_2)
df2
```

```python
# Merged dictionaries
merged_dicts = pd.merge(df1,df2,on=df.make)
merged_dicts
```
